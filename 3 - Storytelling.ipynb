### Converting all string columns (with numbers) to numeric columns

```python
df_yearly['Total_Population19'] = pd.to_numeric(df_yearly['Total_Population19'].str.replace(',',''))
df_yearly['Male populatio 2019'] = pd.to_numeric(df_yearly['Male populatio 2019'].str.replace(',',''))
df_yearly['Female population 2019'] = pd.to_numeric(df_yearly['Female population 2019'].str.replace(',',''))
df_yearly['Households'] = pd.to_numeric(df_yearly['Households'].str.replace(',',''))
df_yearly['LandArea'] = pd.to_numeric(df_yearly['LandArea'].str.replace(',',''))
df_yearly['Population Density'] = pd.to_numeric(df_yearly['Population Density'].str.replace(',',''))
df_yearly['Population in 2009'] = pd.to_numeric(df_yearly['Population in 2009'].str.replace(',',''))
```

```python
df_yearly['Pop_change'] = pd.to_numeric(df_yearly['Pop_change'].str.strip().str.replace('(','').str.replace(')','').str.replace(',',''))
```

```python
# df_yearly.info()
```

```python
def get_perc(pos, pop):
    perc = (pos/pop)*100
    return perc
```

```python
# df_yearly.corr()
df_yearly['Perc_pop_with_HIV'] = np.vectorize(get_perc)(df_yearly['MOH 731   Positive_Total   (Sum  HV01-18 to  HV01-25) HV01-26'], df_yearly['Total_Population19'])
```

```python
df_bar_perc = df_yearly[df_yearly['Perc_pop_with_HIV'] > 0]
df_bar_perc = df_bar_perc.sort_values('Perc_pop_with_HIV', ascending=False)
# df_bar_perc
```

### Insight: 
Counties in the Western and Nyanza Province are most affected by HIV 

```python
fig, ax = plt.subplots(1,1, sharex=True, figsize=(12,6), dpi=200)
fig = sns.barplot(data=df_bar_perc,
                  y='Perc_pop_with_HIV',
                  x='Orgunit name')

fig.set_xticklabels(labels=fig.get_xticklabels(), rotation=90)

ax.set_title('Percentage of population living with HIV (for the counties without missing data)')

plt.show()    
```

```python
df_monthly = df_hiv.groupby('Month').sum().drop('Infection_rate', axis=1)
df_monthly
```

```python
cols = df_monthly.columns
cols
```

```python
months = df_monthly.index
discordant = list(df_monthly[str(cols[0])])
linked = list(df_monthly[str(cols[1])])
neg = list(df_monthly[str(cols[2])])
pos = list(df_monthly[str(cols[3])])
couples = list(df_monthly[str(cols[4])])
tested_new = list(df_monthly[str(cols[5])])
totalhvo1_50 = list(df_monthly[str(cols[6])])
total_tested = list(df_monthly[str(cols[7])])

cols_list = [discordant, linked, neg, pos, couples, tested_new, totalhvo1_50, total_tested]
x = months
```

```python
fig, ax = plt.subplots(1,1,figsize=(8,4), dpi=200)
# ax2 = fig.add_axes([0,0,.25,.25])

ax.plot(months, discordant, label='Discordant')
ax.plot(months, linked, label='Positive last 3 months')
ax.plot(months, pos, label='Tested positive')
ax.plot(months, couples, label='Tested couples')
# ax.set_xlim(1,12)

ax.set_title('Monthly trends for each test type over the year')
ax.set_xlabel('Months')
ax.set_ylabel('Number of people')

plt.legend(title='Legend:', loc=[1.01,0.5])
plt.show()
```

## Conclusion:
1 - Counties with high populations, such as those with big towns or cities, have the highest number of HIV positive patients.
2 - Counties found in the Western and Nyanza Province regions, have higher numbers of HIV positive people.
    This could possibly be attributed to a number of factors. e.g cultural factors such as polygamy, bride inheritance, superstition; and historical factors such as disinformation on HIV and how it spreads.
    
3 - Condoms, and especially male condoms (as they are mostly used), prove effective in curbing the spread of the HIV virus. They should therefore be distributed to high risk counties as much as possible.

4 - Condoms are by far the most popular kind of contraceptive.
